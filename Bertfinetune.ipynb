```python
data_path = '/content/Updated_Nifty50_Dataset.txt'
with open(data_path, 'r') as file:
    contexts = file.readlines()
print(contexts[:10])
```

```python
train_questions_answers = []

# Generate questions for each context
for index, context in enumerate(contexts):
    # Split context to extract relevant details (assuming comma-separated format)
    details = context.split("|")
    company_name = details[0]

    # Define the questions and answers based on the details
    questions_answers = [
        {
            "context_index": index,
            "question": f"What are the sales of {company_name}?",
            "answer": details[1].split(": ")[1]
        },
        {
            "context_index": index,
            "question": f"What are the expenses of {company_name}?",
            "answer": details[2].split(": ")[1]
        },
        {
            "context_index": index,
            "question": f"What is the net profit of {company_name}?",
            "answer": details[3].split(": ")[1]
        },
        {
            "context_index": index,
            "question": f"What are the reserves of {company_name}?",
            "answer": details[4].split(": ")[1]
        },
        {
            "context_index": index,
            "question": f"What are the borrowings of {company_name}?",
            "answer": details[5].split(": ")[1]
        },
        {
            "context_index": index,
            "question": f"What is the promoter holding percentage of {company_name}?",
            "answer": details[6].split(": ")[1]
        },
        {
            "context_index": index,
            "question": f"What is the financial summary for {company_name}?",
            "answer": details[7].split(": ")[1]
        }
    ]

    # Add the generated questions and answers to the main list
    train_questions_answers.extend(questions_answers)

# Save the questions and answers to a file
output_file = "Nifty50_Questions_Answers.json"
import json
with open(output_file, "w") as outfile:
    json.dump(train_questions_answers, outfile, indent=4)

print(f"Questions and answers have been saved to {output_file}")
```

```python
questions_answers_file = "Nifty50_Questions_Answers.json"
with open(questions_answers_file, "r", encoding="utf-8") as file:
    train_questions_answers = json.load(file)
```

```python
# Initialize an empty list to hold the final combined data
final_train_data = []

# Iterate through each context and its corresponding questions and answers
for i, context in enumerate(contexts):
    # Create a list for the questions and answers that belong to this context
    qas = []

    # Filter the questions and answers for the current context (based on context_index)
    for q_and_a in train_questions_answers:
        if q_and_a['context_index'] == i:
            # Clean up spaces and find the answer's start position in the context
            answer_text = q_and_a['answer'].strip()  # Remove leading/trailing spaces
            answer_start = context.find(answer_text)  # Find the start index

            # Create the 'qas' list in the required format
            qas.append({
                "id": str(i+1).zfill(5),  # For example, '00001', '00002', etc.
                "is_impossible": False,   # Assuming answers exist
                "question": q_and_a['question'],
                "answers": [{"text": answer_text, "answer_start": answer_start}]
            })

    # Append the context and its associated questions and answers to the final data
    final_train_data.append({
        "context": context,
        "qas": qas
    })

# Verify the first few entries
print("Sample of Final Data:", final_train_data[:2])
```

```python
# Test contexts
test_contexts = [
    "ICICI Bank Ltd. | Sales: Rs. 150000 Cr | Expense: Rs. 120000 Cr | Net Profit: Rs. 30000 Cr | Reserves: Rs. 75000 Cr | Borrowings: Rs. 45000 Cr | Promoter Holding: 55% | Summary: Risk Present",
    "Hindustan Unilever Ltd. | Sales: Rs. 160000 Cr | Expense: Rs. 128000 Cr | Net Profit: Rs. 32000 Cr | Reserves: Rs. 80000 Cr | Borrowings: Rs. 48000 Cr | Promoter Holding: 56% | Summary: Good Product and Risk Less",
    "State Bank of India | Sales: Rs. 170000 Cr | Expense: Rs. 136000 Cr | Net Profit: Rs. 34000 Cr | Reserves: Rs. 85000 Cr | Borrowings: Rs. 51000 Cr | Promoter Holding: 57% | Summary: Risk Present",
    "Bharti Airtel Ltd. | Sales: Rs. 180000 Cr | Expense: Rs. 144000 Cr | Net Profit: Rs. 36000 Cr | Reserves: Rs. 90000 Cr | Borrowings: Rs. 54000 Cr | Promoter Holding: 58% | Summary: Good Product and Risk Less",
    "Adani Enterprises Ltd. | Sales: Rs. 190000 Cr | Expense: Rs. 152000 Cr | Net Profit: Rs. 38000 Cr | Reserves: Rs. 95000 Cr | Borrowings: Rs. 57000 Cr | Promoter Holding: 59% | Summary: Risk Present"
]

# Initialize an empty list to store formatted data
formatted_data = []

# Loop through each test context
for index, context in enumerate(test_contexts):
    # Split the context into details
    company_info = context.split(" | ")
    company_name = company_info[0].strip()

    sales = company_info[1].split(":")[1].strip()
    expense = company_info[2].split(":")[1].strip()
    net_profit = company_info[3].split(":")[1].strip()
    reserves = company_info[4].split(":")[1].strip()
    borrowings = company_info[5].split(":")[1].strip()
    promoter_holding = company_info[6].split(":")[1].strip()
    summary = company_info[7].split(":")[1].strip()

    # Find the answer start for each piece of information
    sales_start = context.find(sales)
    expense_start = context.find(expense)
    net_profit_start = context.find(net_profit)
    reserves_start = context.find(reserves)
    borrowings_start = context.find(borrowings)
    promoter_holding_start = context.find(promoter_holding)
    summary_start = context.find(summary)

    # Create the question-answer pairs with the correct answer start positions
    questions_answers = [
        {
            "id": f"{index}01",
            "is_impossible": False,
            "question": f"What are the sales of {company_name}?",
            "answers": [{"text": sales, "answer_start": sales_start}]
        },
        {
            "id": f"{index}02",
            "is_impossible": False,
            "question": f"What are the expenses of {company_name}?",
            "answers": [{"text": expense, "answer_start": expense_start}]
        },
        {
            "id": f"{index}03",
            "is_impossible": False,
            "question": f"What is the net profit of {company_name}?",
            "answers": [{"text": net_profit, "answer_start": net_profit_start}]
        },
        {
            "id": f"{index}04",
            "is_impossible": False,
            "question": f"What are the reserves of {company_name}?",
            "answers": [{"text": reserves, "answer_start": reserves_start}]
        },
        {
            "id": f"{index}05",
            "is_impossible": False,
            "question": f"What are the borrowings of {company_name}?",
            "answers": [{"text": borrowings, "answer_start": borrowings_start}]
        },
        {
            "id": f"{index}06",
            "is_impossible": False,
            "question": f"What is the promoter holding percentage of {company_name}?",
            "answers": [{"text": promoter_holding, "answer_start": promoter_holding_start}]
        },
        {
            "id": f"{index}07",
            "is_impossible": False,
            "question": f"What is the financial summary for {company_name}?",
            "answers": [{"text": summary, "answer_start": summary_start}]
        }
    ]

    # Append to the formatted data list
    formatted_data.append({
        "context": context,
        "qas": questions_answers
    })

# Output the formatted data in JSON format for easier viewing
import json
print(json.dumps(formatted_data, indent=2))
```

```python
train_args = {
    'overwrite_output_dir': True,
    "evaluate_during_training": True,
    "max_seq_length": 128,
    "num_train_epochs": 25, #25, after experimentations
    "evaluate_during_training_steps": 500,
    "save_model_every_epoch": False,
    "save_eval_checkpoints": False,
    "n_best_size":16, #batch_size is another important argument
    "train_batch_size": 16,
    "eval_batch_size": 16
}
```

```python
import logging
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
```

```python
model = QuestionAnsweringModel("bert",
                               "bert-large-cased",
                               args = train_args,use_cuda=True)
```

```python
model.train_model(final_train_data, eval_data=formatted_data)
```
